```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sympy as sym
from sklearn.linear_model import LinearRegression
```


#### Variable independiente
experiencia = np.array([[5], [6], [13]]) 

#### Variable objetivo
salario = np.array([6.85, 16.83, 26.84])

```python
x = np.array([[5], [6], [13], [20]])

y = np.array([6.85, 16.83, 26.84, 40])
```

plt.figure(figsize=(8, 6))

```python
plt.figure(figsize=(8, 6))

plt.scatter(x, y, color="gold", s=250,
            marker="o", label="Valor verdadero")
```

```python
modelo = LinearRegression().fit(x.reshape(4, -1), y)
```

plt.figure(figsize=(8, 6))

plt.scatter(experiencia, modelo.predict(experiencia),
            color="blue", s=250, marker="P", label="Valor predicho") 

```python
plt.figure(figsize=(8, 6))

plt.scatter(x, modelo.predict(x),
            color="blue", s=250, marker="P", label="Valor predicho") 
```
